```python
# ============================================================================
# IMPORTS
# ============================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import warnings
warnings.filterwarnings('ignore')

print(f"TensorFlow Version: {tf.__version__}")
print(f"GPU Available: {tf.config.list_physical_devices('GPU')}")

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# ============================================================================
# 1. LOAD AND PREPARE DATA
# ============================================================================
print("="*80)
print("LOADING DATASET")
print("="*80)

df = pd.read_csv("D:\\10 sem\\novintix\\Wind Turbine\\T1.csv") 
df['Date/Time'] = pd.to_datetime(df['Date/Time'], format='%d %m %Y %H:%M')
df = df.sort_values('Date/Time').reset_index(drop=True)

print(f"Dataset Shape: {df.shape}")
print(f"Date Range: {df['Date/Time'].min()} to {df['Date/Time'].max()}")

# ============================================================================
# 2. CREATE WINDOWED DATASET FUNCTION
# ============================================================================
def create_windowed_dataset(data, window_size=10, forecast_horizon=1):
    """
    Creates windowed time-series data for LSTM training
    
    Args:
        data: numpy array of time series data
        window_size: number of past timesteps to use for prediction
        forecast_horizon: number of future timesteps to predict
    
    Returns:
        X: input sequences (samples, window_size, features)
        y: target values (samples, forecast_horizon)
    """
    X, y = [], []
    
    for i in range(len(data) - window_size - forecast_horizon + 1):
        X.append(data[i:i + window_size])
        y.append(data[i + window_size:i + window_size + forecast_horizon])
    
    return np.array(X), np.array(y)

# ============================================================================
# 3. PREPARE DATA FOR EACH VARIABLE
# ============================================================================
# Parameters
WINDOW_SIZE = 24  # Use past 24 timesteps (4 hours of data at 10-min intervals)
FORECAST_HORIZON = 6  # Predict next 6 timesteps (1 hour ahead)
TRAIN_SPLIT = 0.8

# Variables to forecast
variables = {
    'LV ActivePower (kW)': 'Active Power',
    'Wind Speed (m/s)': 'Wind Speed',
    'Theoretical_Power_Curve (KWh)': 'Theoretical Power',
    'Wind Direction (°)': 'Wind Direction'
}

# Dictionary to store results
results = {}

print("\n" + "="*80)
print(f"TIME-SERIES CONFIGURATION")
print("="*80)
print(f"Window Size: {WINDOW_SIZE} timesteps (4 hours)")
print(f"Forecast Horizon: {FORECAST_HORIZON} timesteps (1 hour ahead)")
print(f"Train/Test Split: {TRAIN_SPLIT*100:.0f}% / {(1-TRAIN_SPLIT)*100:.0f}%")

# ============================================================================
# 4. BUILD AND TRAIN MODELS FOR EACH VARIABLE
# ============================================================================
for col, name in variables.items():
    print("\n" + "="*80)
    print(f"TRAINING MODEL FOR: {name}")
    print("="*80)
    
    # Extract data
    data = df[col].values.reshape(-1, 1)
    
    # Scale data to [0, 1] range
    scaler = MinMaxScaler(feature_range=(0, 1))
    data_scaled = scaler.fit_transform(data)
    
    # Create windowed dataset
    X, y = create_windowed_dataset(data_scaled, WINDOW_SIZE, FORECAST_HORIZON)
    
    # Split into train and test
    train_size = int(len(X) * TRAIN_SPLIT)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]
    
    print(f"Training samples: {len(X_train)}")
    print(f"Testing samples: {len(X_test)}")
    print(f"Input shape: {X_train.shape}")
    print(f"Output shape: {y_train.shape}")
    
    # Build LSTM model
    model = Sequential([
        LSTM(128, activation='relu', return_sequences=True, 
             input_shape=(WINDOW_SIZE, 1)),
        Dropout(0.2),
        LSTM(64, activation='relu', return_sequences=True),
        Dropout(0.2),
        LSTM(32, activation='relu'),
        Dropout(0.2),
        Dense(FORECAST_HORIZON)
    ])
    
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='mse',
        metrics=['mae']
    )
    
    print("\nModel Architecture:")
    model.summary()
    
    # Callbacks
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        verbose=1
    )
    
    reduce_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=0.00001,
        verbose=1
    )
    
    # Train model
    print("\nTraining model...")
    history = model.fit(
        X_train, y_train,
        epochs=50,
        batch_size=32,
        validation_split=0.2,
        callbacks=[early_stopping, reduce_lr],
        verbose=1
    )
    
    # Make predictions
    print("\nMaking predictions...")
    y_train_pred = model.predict(X_train, verbose=0)
    y_test_pred = model.predict(X_test, verbose=0)
    
    # Inverse transform predictions to original scale
    y_train_actual = scaler.inverse_transform(y_train.reshape(-1, 1)).flatten()
    y_train_pred_rescaled = scaler.inverse_transform(y_train_pred.reshape(-1, 1)).flatten()
    
    y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()
    y_test_pred_rescaled = scaler.inverse_transform(y_test_pred.reshape(-1, 1)).flatten()
    
    # Calculate metrics
    train_rmse = np.sqrt(mean_squared_error(y_train_actual, y_train_pred_rescaled))
    train_mae = mean_absolute_error(y_train_actual, y_train_pred_rescaled)
    train_r2 = r2_score(y_train_actual, y_train_pred_rescaled)
    
    test_rmse = np.sqrt(mean_squared_error(y_test_actual, y_test_pred_rescaled))
    test_mae = mean_absolute_error(y_test_actual, y_test_pred_rescaled)
    test_r2 = r2_score(y_test_actual, y_test_pred_rescaled)
    
    # Calculate MAPE (Mean Absolute Percentage Error)
    train_mape = np.mean(np.abs((y_train_actual - y_train_pred_rescaled) / 
                                 (y_train_actual + 1e-10))) * 100
    test_mape = np.mean(np.abs((y_test_actual - y_test_pred_rescaled) / 
                                (y_test_actual + 1e-10))) * 100
    
    print("\n" + "-"*80)
    print("EVALUATION METRICS")
    print("-"*80)
    print(f"{'Metric':<20} {'Train':<15} {'Test':<15}")
    print("-"*80)
    print(f"{'RMSE':<20} {train_rmse:<15.4f} {test_rmse:<15.4f}")
    print(f"{'MAE':<20} {train_mae:<15.4f} {test_mae:<15.4f}")
    print(f"{'R² Score':<20} {train_r2:<15.4f} {test_r2:<15.4f}")
    print(f"{'MAPE (%)':<20} {train_mape:<15.4f} {test_mape:<15.4f}")
    print("-"*80)
    
    # Store results
    results[name] = {
        'model': model,
        'scaler': scaler,
        'history': history,
        'y_train_actual': y_train_actual,
        'y_train_pred': y_train_pred_rescaled,
        'y_test_actual': y_test_actual,
        'y_test_pred': y_test_pred_rescaled,
        'metrics': {
            'train_rmse': train_rmse,
            'train_mae': train_mae,
            'train_r2': train_r2,
            'train_mape': train_mape,
            'test_rmse': test_rmse,
            'test_mae': test_mae,
            'test_r2': test_r2,
            'test_mape': test_mape
        }
    }
    
    # Save model
    model.save(f'model_{name.lower().replace(" ", "_")}.h5')
    print(f"\n✓ Model saved: model_{name.lower().replace(' ', '_')}.h5")

# ============================================================================
# 5. VISUALIZE TRAINING HISTORY
# ============================================================================
print("\n" + "="*80)
print("GENERATING TRAINING HISTORY PLOTS")
print("="*80)

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Training History - Loss and MAE', fontsize=16, fontweight='bold')

for idx, (name, result) in enumerate(results.items()):
    row = idx // 2
    col = idx % 2
    ax = axes[row, col]
    
    history = result['history']
    
    ax.plot(history.history['loss'], label='Train Loss', linewidth=2)
    ax.plot(history.history['val_loss'], label='Val Loss', linewidth=2)
    ax.set_xlabel('Epoch', fontweight='bold')
    ax.set_ylabel('Loss (MSE)', fontweight='bold')
    ax.set_title(f'{name} - Training History', fontweight='bold')
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('task2_training_history.png', dpi=300, bbox_inches='tight')
print("✓ Saved: task2_training_history.png")
plt.show()

# ============================================================================
# 6. VISUALIZE PREDICTIONS VS ACTUAL
# ============================================================================
print("\n" + "="*80)
print("GENERATING PREDICTION PLOTS")
print("="*80)

fig, axes = plt.subplots(4, 2, figsize=(18, 16))
fig.suptitle('Time-Series Forecasting: Predicted vs Actual Values', 
             fontsize=16, fontweight='bold')

for idx, (name, result) in enumerate(results.items()):
    # Train predictions
    ax_train = axes[idx, 0]
    sample_size_train = min(500, len(result['y_train_actual']))
    ax_train.plot(result['y_train_actual'][:sample_size_train], 
                  label='Actual', linewidth=1.5, alpha=0.7)
    ax_train.plot(result['y_train_pred'][:sample_size_train], 
                  label='Predicted', linewidth=1.5, alpha=0.7)
    ax_train.set_xlabel('Sample Index', fontweight='bold')
    ax_train.set_ylabel('Value', fontweight='bold')
    ax_train.set_title(f'{name} - Training Set\n'
                       f'RMSE: {result["metrics"]["train_rmse"]:.2f}, '
                       f'R²: {result["metrics"]["train_r2"]:.3f}',
                       fontweight='bold')
    ax_train.legend()
    ax_train.grid(True, alpha=0.3)
    
    # Test predictions
    ax_test = axes[idx, 1]
    sample_size_test = min(500, len(result['y_test_actual']))
    ax_test.plot(result['y_test_actual'][:sample_size_test], 
                 label='Actual', linewidth=1.5, alpha=0.7)
    ax_test.plot(result['y_test_pred'][:sample_size_test], 
                 label='Predicted', linewidth=1.5, alpha=0.7)
    ax_test.set_xlabel('Sample Index', fontweight='bold')
    ax_test.set_ylabel('Value', fontweight='bold')
    ax_test.set_title(f'{name} - Test Set\n'
                      f'RMSE: {result["metrics"]["test_rmse"]:.2f}, '
                      f'R²: {result["metrics"]["test_r2"]:.3f}',
                      fontweight='bold')
    ax_test.legend()
    ax_test.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('task2_predictions_vs_actual.png', dpi=300, bbox_inches='tight')
print("✓ Saved: task2_predictions_vs_actual.png")
plt.show()

# ============================================================================
# 7. SCATTER PLOTS (Predicted vs Actual)
# ============================================================================
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
fig.suptitle('Prediction Accuracy - Scatter Plots (Test Set)', 
             fontsize=16, fontweight='bold')

for idx, (name, result) in enumerate(results.items()):
    row = idx // 2
    col = idx % 2
    ax = axes[row, col]
    
    # Scatter plot
    ax.scatter(result['y_test_actual'], result['y_test_pred'], 
               alpha=0.5, s=10)
    
    # Perfect prediction line
    min_val = min(result['y_test_actual'].min(), result['y_test_pred'].min())
    max_val = max(result['y_test_actual'].max(), result['y_test_pred'].max())
    ax.plot([min_val, max_val], [min_val, max_val], 
            'r--', linewidth=2, label='Perfect Prediction')
    
    ax.set_xlabel('Actual Values', fontweight='bold')
    ax.set_ylabel('Predicted Values', fontweight='bold')
    ax.set_title(f'{name}\nR² = {result["metrics"]["test_r2"]:.3f}', 
                 fontweight='bold')
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('task2_scatter_plots.png', dpi=300, bbox_inches='tight')
print("✓ Saved: task2_scatter_plots.png")
plt.show()

# ============================================================================
# 8. METRICS COMPARISON TABLE
# ============================================================================
print("\n" + "="*80)
print("FINAL METRICS COMPARISON")
print("="*80)

metrics_df = pd.DataFrame({
    'Variable': list(results.keys()),
    'Train RMSE': [r['metrics']['train_rmse'] for r in results.values()],
    'Test RMSE': [r['metrics']['test_rmse'] for r in results.values()],
    'Train MAE': [r['metrics']['train_mae'] for r in results.values()],
    'Test MAE': [r['metrics']['test_mae'] for r in results.values()],
    'Train R²': [r['metrics']['train_r2'] for r in results.values()],
    'Test R²': [r['metrics']['test_r2'] for r in results.values()],
    'Test MAPE (%)': [r['metrics']['test_mape'] for r in results.values()]
})

print(metrics_df.to_string(index=False))

# Save metrics to CSV
metrics_df.to_csv('task2_metrics_summary.csv', index=False)
print("\n✓ Saved: task2_metrics_summary.csv")

# ============================================================================
# 9. SUMMARY REPORT
# ============================================================================
print("\n" + "="*80)
print("TASK 2 - TIME-SERIES FORECASTING SUMMARY")
print("="*80)
print(f"""
Configuration:
  • Window Size: {WINDOW_SIZE} timesteps (4 hours)
  • Forecast Horizon: {FORECAST_HORIZON} timesteps (1 hour ahead)
  • Train/Test Split: {TRAIN_SPLIT*100:.0f}%/{(1-TRAIN_SPLIT)*100:.0f}%
  • Model Architecture: LSTM with 3 layers (128→64→32 units)
  • Training Samples: {train_size:,}
  • Testing Samples: {len(X_test):,}

Model Performance:
""")

for name, result in results.items():
    print(f"\n  {name}:")
    print(f"    Test RMSE: {result['metrics']['test_rmse']:.4f}")
    print(f"    Test MAE:  {result['metrics']['test_mae']:.4f}")
    print(f"    Test R²:   {result['metrics']['test_r2']:.4f}")
    print(f"    Test MAPE: {result['metrics']['test_mape']:.2f}%")

print(f"""
Deliverables Generated:
  ✓ task2_training_history.png
  ✓ task2_predictions_vs_actual.png
  ✓ task2_scatter_plots.png
  ✓ task2_metrics_summary.csv
  ✓ 4 trained models (.h5 files)

All models have been trained and evaluated successfully!
""")
print("="*80)
print("TASK 2 COMPLETED SUCCESSFULLY!")
print("="*80)
```
